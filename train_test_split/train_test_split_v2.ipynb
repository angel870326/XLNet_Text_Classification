```
import numpy as np
import pandas as pd
from collections import Counter
from sklearn.model_selection import train_test_split
```

```
from google.colab import drive
drive.mount('/content/drive')
```

## **品類**


---



```
# Raw dataset
product_train_raw = pd.read_csv('/content/drive/MyDrive/碩一下/金融科技/麻布數據/第四組/資料集/category_clean1.csv')
product_train_raw = product_train_raw.dropna()
product_train_raw.head()
```

```
# Clean dataset
product_train_data = {'id': range(1, len(product_train_raw) + 1),
              'name': product_train_raw['name'],
              'product': product_train_raw['product']}
product_train = pd.DataFrame(product_train_data)
product_train
```

```
# product counter
product_count = Counter(product_train['product'].values)
product_count
```

```
# Train test split
product_training, product_testing = train_test_split(product_train, test_size=0.2, random_state=42)
product_training
```

```
product_training.to_csv('/content/drive/MyDrive/碩一下/金融科技/麻布數據/第四組/資料集/category_train.csv', index=False)
product_testing.to_csv('/content/drive/MyDrive/碩一下/金融科技/麻布數據/第四組/資料集/category_test.csv', index=False)
```
