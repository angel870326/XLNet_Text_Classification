# **品類過採樣（Oversampling for Product）**

### **Contents:**

*   清理後的資料
*   區分出大於30筆&小於30筆的資料
*   **小於30筆者過採樣(oversampling)至100筆***
*   合併


```
import numpy as np
import pandas as pd
from collections import Counter
import re
```

```
from google.colab import drive
drive.mount('/content/drive')
```

## **清理後的資料**

```
# Cleaned dataset
product_train_or = pd.read_csv('/content/drive/MyDrive/碩一下/金融科技/麻布數據/第四組/資料集/category_clean1.csv')
product_train_or
```

```
# product counter
product_train_or['counter'] = product_train_or.groupby('product')['product'].transform('count')
product_train_or
```

## **區分出大於30筆&小於30筆的資料**

```
# 大於30筆
product_train_large = product_train_or[product_train_or.counter > 30]
product_train_large
```

```
# 小於30筆
product_train_small = product_train_or[product_train_or.counter <= 30]
product_train_small
```

```
# 小於30筆的品類
product_count = Counter(product_train_small['product'].values)
product_count_df = pd.DataFrame(list(product_count.items()), columns=['Product', 'Count']).sort_values(by=['Count'])
product_count_df
```

## **小於30筆者過採樣(oversampling)至100筆**

```
from sklearn.utils import resample
filter_1 = product_train_small["product"].isin([product_count_df.iloc[0,0]])
product_1 = product_train_small.loc[filter_1,:]
oversample = resample(product_1,
           replace = True,
           n_samples = 100,
           random_state = 42)

for i in range(1, product_count_df.shape[0]):
  filter_i = product_train_small["product"].isin([product_count_df.iloc[i,0]])
  product_i = product_train_small.loc[filter_i,:]
  oversample_i = resample(product_i,
             replace = True,
             n_samples = 100,
             random_state = 42)
  oversample = pd.concat([oversample, oversample_i])

oversample
```

## **合併**

```
product_train = pd.concat([oversample, product_train_large])
product_train
```

```
product_train.to_csv('/content/drive/MyDrive/碩一下/金融科技/麻布數據/第四組/資料集/category_clean1_oversample.csv', index = False)
```
