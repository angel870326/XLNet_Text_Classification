## **XLNet v.s. BERT 0.85**


---



```
brand_concat85 = pd.concat([brand_xlnet, brand_meng_85['brand']], axis=1)
brand_concat85
```

```
comparison = np.where(brand_concat85["brand_predict"] == brand_concat85["brand"], True, False)
brand_concat85['equal'] = comparison
brand_concat85
```

### 預測結果相同的

```
brand_equal85 = brand_concat85[brand_concat85['equal'] == True]
brand_equal85
```

```
xlnet_equal85_under1 = brand_equal85[brand_equal85['max_prob']<0.1]
xlnet_equal85_under1
```

### 預測結果不同的

```
brand_unequal85 = brand_concat85[brand_concat85['equal'] == False]
brand_unequal85
```

```
brand_unequal85_notother = brand_unequal85[brand_unequal85['brand']!='Other']
brand_unequal85_notother
```

```
xlnet_unequal85_over5 = brand_unequal85_notother[brand_unequal85_notother['max_prob']>0.5]
xlnet_unequal85_over5
```

```
xlnet_unequal85_over5.to_excel('/content/drive/MyDrive/碩一下/金融科技/麻布數據/第四組/比較與統整/xlnet_unequal85_over5.xlsx')
```

## **XLNet v.s. BERT 0.85~0.5**


---



```
bert_concat = pd.concat([brand_xlnet, brand_meng_85['brand'], brand_meng_50['brand']], axis=1)
bert_concat.columns = ['id', 'name', 'max_prob', 'brand', '85', '50']
bert_concat
```

```
comparison_bert = np.where(bert_concat['85'] == bert_concat['50'], True, False)
bert_concat['equal'] = comparison_bert
bert_concat
```

```
bert_85to50 = bert_concat[bert_concat['equal'] == False]
bert_85to50
```

```
comparison_xlnet_bert50 = np.where(bert_85to50['brand'] == bert_85to50['50'], True, False)
bert_85to50['equal'] = comparison_xlnet_bert50
bert_85to50
```

### 預測結果相同的

```
bert_85to50_equal = bert_85to50[bert_85to50['equal'] == True]
bert_85to50_equal
```

```
bert_85to50_equal.to_excel('/content/drive/MyDrive/碩一下/金融科技/麻布數據/第四組/比較與統整/bert_85to50_equal.xlsx')
```

### 預測結果不同的

```
bert_85to50_unequal = bert_85to50[bert_85to50['equal'] == False]
bert_85to50_unequal
```

```
xlnet_85to50_unequal_over5 = bert_85to50_unequal[bert_85to50_unequal['max_prob']>0.5]
xlnet_85to50_unequal_over5
```

```
xlnet_85to50_unequal_over5.to_excel('/content/drive/MyDrive/碩一下/金融科技/麻布數據/第四組/比較與統整/xlnet_85to50_unequal_over5.xlsx')
```
