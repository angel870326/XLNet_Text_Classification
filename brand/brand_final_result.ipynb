## **更新原有品牌清單更名者**

```
def new_brand_name(df, col_name):
  df[col_name] = df[col_name].replace(['tide','naturalsbywatsons','more','家福carrefour','大拇指fp','luna'],['汰漬tide','屈臣氏watsons','多喝水','家樂福','大潤發','佳麗寶kanebo'])
```

## **品牌統一清單**

```
brand_list = pd.read_csv('/content/drive/MyDrive/碩一下/金融科技/麻布數據/第四組/品牌/品牌清單/brand_dict_final.csv')
brand_list.columns = ['id', 'original', 'organized']
brand_list
```

```
original_brand = brand_list["original"].to_numpy()
organized_brand = brand_list["organized"].to_numpy()
original_brand
```

```
def update_brand_list(df, col_name):
  df[col_name] = df[col_name].replace(original_brand, organized_brand)
```

## **XLNet**


---



```
brand_xlnet = pd.read_csv('/content/drive/MyDrive/碩一下/金融科技/麻布數據/第四組/output/final/xlnet_brand_single_predict_bcel.csv')
new_brand_name(brand_xlnet, 'brand_predict')
update_brand_list(brand_xlnet, 'brand_predict')
brand_xlnet
```

## **BERT 0.85以上**

---
by [Meng-Chieh, Liu](https://github.com/MengChiehLiu)


```
brand_result = pd.read_csv('/content/drive/MyDrive/碩一下/金融科技/麻布數據/第四組/孟頡/test_data_0.85.csv')
brand_result.columns = ['id', 'name', 'brand']
new_brand_name(brand_result, 'brand')
update_brand_list(brand_result, 'brand')
brand_result
```

## **Web Craw New Brand**


---
by [Meng-Chieh, Liu](https://github.com/MengChiehLiu)



```
webcraw = pd.read_csv('/content/drive/MyDrive/碩一下/金融科技/麻布數據/第四組/孟頡/new_brands_v3.csv')
webcraw.columns = ['id', 'name', 'new_brand']
new_brand_name(webcraw, 'new_brand')
update_brand_list(webcraw, 'new_brand')
webcraw
```

```
for i in webcraw.index:
  brand_result.loc[brand_result['id'] == webcraw.loc[i,'id'], 'brand'] = webcraw.loc[i,'new_brand']
brand_result
```

## **品牌更新答案清單**


---

XLNet 0.5 以上預測正確者


```
modify_list = pd.read_excel("/content/drive/MyDrive/碩一下/金融科技/麻布數據/第四組/比較與統整/品牌/brand_modify_list.xlsx")
modify_list = modify_list.astype({"id": int})
new_brand_name(modify_list, 'brand')
update_brand_list(modify_list, 'brand')
modify_list
```

## **沒有找到的新品牌以 XLNet 預測結果替代**


---



```
zero_filter = brand_result["brand"].isin(["0","Other","other"])
zero = brand_result.loc[zero_filter,:]
zero
```

```
for i in zero.index:
  brand_result.loc[i, 'brand'] = brand_xlnet.loc[i,'brand_predict']
brand_result
```
