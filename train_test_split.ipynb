```
import numpy as np
import pandas as pd
from collections import Counter
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

## **品類**


---



```
# Raw dataset
product_train_raw = pd.read_csv('/content/drive/MyDrive/碩一下/金融科技/麻布數據/資料檔/品類_clean.csv')
product_train_raw = product_train_raw.dropna()
product_train_raw.head()
```

```
# Clean dataset
product_train_data = {'id': range(1, len(product_train_raw) + 1),
              'name': product_train_raw['name'],
              'clean_name': product_train_raw['clean_name'],
              'product': product_train_raw['product']}
product_train = pd.DataFrame(product_train_data)
product_train
```

```
# product counter
product_count = Counter(product_train['product'].values)
product_count
```

```
# Train test split
product_training, product_testing = train_test_split(product_train, test_size=0.2, random_state=42)
product_training
```

```
product_training.to_csv('/content/drive/MyDrive/碩一下/金融科技/麻布數據/資料檔/category_train.csv', index=False)
product_testing.to_csv('/content/drive/MyDrive/碩一下/金融科技/麻布數據/資料檔/category_test.csv', index=False)
```

## **品牌**


---



```
# Raw dataset
brand_train_raw = pd.read_csv('/content/drive/MyDrive/碩一下/金融科技/麻布數據/資料檔/品牌_clean.csv')
brand_train_raw = brand_train_raw.dropna()
brand_train_raw.head()
```

```
# Clean dataset
brand_train_data = {'id': range(1, len(brand_train_raw) + 1),
              'name': brand_train_raw['name'],
              'brand': brand_train_raw['brand'],
              'clean_name': brand_train_raw['clean_name'],
              'clean_brand': brand_train_raw['clean_brand']}
brand_train = pd.DataFrame(brand_train_data)
brand_train
```

```
# brand counter
brand_count = Counter(brand_train['brand'].values)
brand_count
```

```
# Train test split
brand_training, brand_testing = train_test_split(brand_train, test_size=0.2, random_state=42)
brand_training
```

```
brand_training.to_csv('/content/drive/MyDrive/碩一下/金融科技/麻布數據/資料檔/brand_train.csv', index=False)
brand_testing.to_csv('/content/drive/MyDrive/碩一下/金融科技/麻布數據/資料檔/brand_test.csv', index=False)
```
